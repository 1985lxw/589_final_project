```python
from sklearn import datasets
from tiffany_nn import NN, create_weights, kfold
import pandas as pd
%load_ext autoreload
%autoreload 2
```

```python
digits = datasets.load_digits(return_X_y=True)
digits_dataset_X = digits[0]
digits_dataset_y = digits[1]
N = len(digits_dataset_X)
```

```python
df = pd.DataFrame(digits_dataset_X)
df['label'] = digits_dataset_y
df
```

```python
from itertools import product

layer_configs = [
    [32],
    [64],
    [64, 32],
    [32, 16],
    [128, 64],
    [64, 32, 16],
    [128, 64, 32]
]

lambdas = [0.1, 1e-2, 1e-4]
for layers, l in product(layer_configs, lambdas):
    weights = create_weights([len(df.columns)-1, *layers, 1])
    model = NN(weights=weights, epochs=100, l=l, alpha=0.1)
    metrics = kfold(df, 10, model)
    print(f"Layer Configuration: {layers}, Lambda: {l}")
    print(f"Accuracy: {metrics[0]}, Precision: {metrics[1]}, Recall: {metrics[2]}, F1 Score: {metrics[3]}")
    print("\n")
```
